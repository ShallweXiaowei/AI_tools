```python
import requests

class OllamaChat:
    def __init__(self, model="deepseek-r1:32b"):
        self.model = model
        self.history = []

    def chat(self, user_input):
        self.history.append({"role": "user", "content": user_input})

        # æ ¼å¼åŒ– promptï¼ˆç®€å•æ‹¼æ¥å†å²ï¼‰
        prompt = ""
        for turn in self.history:
            role = turn["role"]
            content = turn["content"]
            if role == "user":
                prompt += f"ç”¨æˆ·: {content}\n"
            else:
                prompt += f"åŠ©æ‰‹: {content}\n"
        prompt += "åŠ©æ‰‹:"

        # è¯·æ±‚ Ollama
        response = requests.post(
            "http://localhost:11434/api/generate",
            json={
                "model": self.model,
                "prompt": prompt,
                "stream": False
            }
        )
        result = response.json()
        reply = result["response"].strip()

        self.history.append({"role": "assistant", "content": reply})
        return reply

# ç¤ºä¾‹ä½¿ç”¨
chatbot = OllamaChat()

while True:
    user_input = input("ä½ : ")
    if user_input.lower() in ["exit", "quit"]:
        break
    response = chatbot.chat(user_input)
    print("ğŸ¤–:", response)
```
